**1. Text cleaning** <br>
Text cleaning should also be performed before preprocessing text. <br>
This can include removing punctuation and replacing wrong characters in the text.

```python
text = "The quick brown fox is running with the lazyyy dog. Dogs are great pets."

## Clean text by removing uppercase, punctuation and cleaning lazyyy
clean_text = text.lower().replace(".", "").replace("yyy", "y")
clean_text
```

**2. Tokenization with nltk** <br>
Let's tokenize the input using ntlk's `word_tokenize()` function.

```python
nltk.download("punkt_tab")
```

```python
## Tokenization
tokens = word_tokenize(clean_text, preserve_line=True)
print("Tokens:", tokens)
```

**3. Stopword removal** <br>
Stopwords are words that have little to no significant meaning in text ("the," "is," "a," "an," "in"). <br>
Removing stopwords can help reduce the vocabulary size (number of unique words) of our text.


```python
## Removal of stopwords
stop_words = stopwords.words("english")
filtered_tokens = [t for t in tokens if t not in stop_words]

print("Original tokens:", tokens)
print("Filtered tokens:", filtered_tokens)
```

**4. Lemmatization/Stemming** <br>
Once stop words have been removed from the text, we can apply lemmatization or stemming

```python
## Lemmatization  -> slower, more accurate
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
print("Lemmatized Tokens:", lemmatized_tokens)
```

```python
## Stemming  -> faster; but may not be a real word
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]
print("Stemmed Tokens:", stemmed_tokens)
```

The example demonstrates the complete pipeline of tokenization, stopword removal, lemmatization, and stemming.<br>  Once your text is suffiently processed, you can convert this string data into **numerical feature vectors** to be fed into a model.

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer

nltk.download("stopwords")
nltk.download("averaged_perceptron_tagger")
nltk.download("wordnet")
nltk.download("averaged_perceptron_tagger_eng")
```


**1. Load the text file with `open()` and `.load()`**

```python
file_name = "nlp.txt"
with open(file_name, "r") as file:
    content = file.read()

print(content[:100])
```

**2. Convert the loaded file to a string object and split the sentences with `.` into a list (`.split`)**

```python
sentences = content.split(".")
print(sentences)
```

**3. Write a loop statement that cleans each sentence of the text by:**
  - Removing extra white spaces at the beginning
  - Removing the extra `##` characters
  - Removing capital letters from words
  - Removing punctuation (use the `'[^\w\s]'` regex pattern)


```python
rem_punctuation_regex_pattern = r"[^\w\s]"
clean_sentences = [
    re.sub(
        rem_punctuation_regex_pattern, "", sentence.strip().lower().replace("##", "")
    )
    for sentence in sentences
]
clean_sentences
```

**4. Create a function to preprocess your text with tokenization, removal of stopwords and lemmatization** with nltk <br>
Apply the function to the list of clean sentences.

```python
# Tokenization
tokens = word_tokenize(" ".join(c for c in clean_sentences), preserve_line=True)
print(len(tokens), "Tokens", tokens)
## Removal of stopwords
stop_words = stopwords.words("english")
filtered_tokens = [t for t in tokens if t not in stop_words]
print(len(filtered_tokens), "filtered tokens", filtered_tokens)
## Lemmatization  -> slower, more accurate
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
print(len(lemmatized_tokens), "Lemmatized Tokens:", lemmatized_tokens)
```

```python
def clean_to_lematize_pipeline(raw_text: str) -> list[str]:
    """
    Returns the lemmatized tokens form the raw text.
    """
    clean_text = re.sub(
        rem_punctuation_regex_pattern, "", raw_text.strip().lower().replace("##", "")
    )
    tokens = [
        token
        for token in word_tokenize(clean_text, preserve_line=True)
        if token not in stopwords.words("english")
    ]
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word=token) for token in tokens]
    return lemmatized_tokens
```

```python
lemmatized_tokens = clean_to_lematize_pipeline(content)
```

**5. Apply POS tagging to the processed tokens with `nltk.pos_tag()`** <br>
The POS tags are abbreviated as follows:
- `NNP`: Proper noun, singular
- `VB`: Verb (VBD: past tense, VBG: present tense, ...)
- `DT`: Determiner
- `JJ`: Adjective
- `NN`: Noun, singular or mass
- `IN`: Preposition or subordinating conjunction
- `DS`: Cardinal number


You can find more information on `pos_tag()` [here](https://www.nltk.org/book/ch05.html).

```python
pos_tag_tokens = nltk.pos_tag(lemmatized_tokens)
pos_tag_tokens
```

**6. Count the occurence of each POS's tag in the whole text** (not on a sentence elevel). <br>
Then, plot the occurence of each POS tag** using `sns.barplot()`

**1. Color-space conversion**: <br>

OpenCV uses an BGR image format to process images. <br>

- OpenCV color space conversion with `cv2.cvtColor()`

```python
# Image in RGB

rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print("RGB grid shape:", rgb_image.shape, "\n")

plt.figure(figsize=(5, 4))
plt.imshow(rgb_image)
plt.title("RGB Image")
plt.axis("off")
plt.show()
```

The grayscale image is represented by a single matrix with shape 961x1683 (size in pixel of the image). <br>

**2. Thresholding** <br>

Thresholding assigns the pixels of an image to True or False, based on the pixel's value, location or both. <br>
We will apply a binary threshold to our grayscale image using `cv2.threshold(src,thresh,maxval,type)`
- `src`: Input image with a single channel (grayscale for example) <br>
- `threshold`: Threshold value.
- `maxval`: Maximum value to be assigned to the pixels that pass the threshold.
- `type`: Thresholding technique to be applied (in our case `cv2.THRESH_BINARY`)

```python
# Thresholding
_, thresholded_image = cv2.threshold(
    gray_image, 200, 255, cv2.THRESH_BINARY
)  # 255 => white pixel

plt.figure(figsize=(12, 10))
plt.subplot(1, 2, 1)
plt.imshow(thresholded_image, cmap="gray")
plt.title("Thresholded Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap="gray")
plt.title("Original grayscale Image")
plt.axis("off")

plt.show()
```

Contour detection detects the borders of objects in an image with `cv2.findContours()`. <br>

**2. Line detection**

Line detection refers to the process of identifying straight lines or linear structures within an image. <br>
Real-life applications of Line detection include :
- **Lane Detection in Autonomous Vehicles**: Assist in autonomous navigation and lane departure warning systems
- **Document Scanning**: Identify the edges of a paper to straighten an image
- **Cartography**: Extract roads, rivers from satellite or aerial imagery to create maps

<br>

<img src = https://i.stack.imgur.com/IcDQ6.png width = "350" height = "220" >


We will apply line detection using the **Hough Line Transform** with `cv2.HoughLinesP()`. <br>
The Hough Line Transform represents straight lines as a collection of points.

- Lines are represented in polar coordinate system:  $ r_{\theta} = x_0 \cos \theta + y_0 \sin \theta $  <br> with $\rho$ the shortest distance from the origin to the line and $\theta$ the angle between the x_axis and the distance line  

- Detect lines by finding the ($\rho$, $\theta$) pairs that have a number of intersections larger than a certain threshold. <br> These intersections represent the points that belong to a same line

<br>

<img src = https://media.geeksforgeeks.org/wp-content/uploads/line-detection-1.png width = "250" height = "200" > <br>






The Hough Line transform is a complex algorithm that isn't easy to explain with a few sentences. <br>
You can learn more about it reading this [Medium article](https://medium.com/@tomasz.kacmajor/hough-lines-transform-explained-645feda072ab) or going to opencv's [website](https://docs.opencv.org/3.4/d9/db0/tutorial_hough_lines.html).

```python
# Line detection
lines = cv2.HoughLinesP(
    image=thresholded_image,
    rho=1,
    theta=np.pi / 180,
    threshold=100,  # minimum number of intersections to identify a line
    minLineLength=100,  # minimun pixel length of a line
    maxLineGap=10,
)  # maximal gap between segments to be connected to a line
```

```python
# Draw the lines of the image
line_image = np.zeros_like(image)
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(line_image, (x1, y1), (x2, y2), (255, 0, 0), 2)
```

```python
# Display the line image
plt.figure()
plt.imshow(cv2.cvtColor(line_image, cv2.COLOR_BGR2RGB))
# plt.imshow(line_image)
plt.title("Line Image")
plt.axis("off")
plt.show()
```

```python
# 1. Read txt file
path = "nlp.txt"

with open(path, "r") as f:
    text = f.read()
```

```python
# 2. Convert str object and split into sentences
text_list = str(text).split(".")
text_list
```

```python
# 3. Clean each sentence of the given text
clean_text = []
for sentence in text_list:
    sentence = sentence.strip()
    sentence = sentence.replace("##", "")
    sentence = sentence.lower()
    sentence = re.sub(r"[^\w\s]", "", sentence)
    clean_text.append(sentence)
```

```python
clean_text
```

```python
# 4. Preprocess text with tokenization, removal stopwords and lemmatization
def text_preprocessing_pipeline(list_text):
    process_tokens = []
    for text in list_text:
        tokens = word_tokenize(text, preserve_line=True)
        filtered_tokens = [t for t in tokens if t not in stop_words]
        lem_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
        process_tokens.append(lem_tokens)

    return process_tokens


# apply preprocessing pipline
tokens = text_preprocessing_pipeline(clean_text)
tokens
```

```python
# 5. Apply POS tagging with nltk
pos_tag = [nltk.pos_tag(t) for t in tokens]
```

```python
# 6. Count the occurence of each POS's tag in the text
list_pos_tags = [token[1] for elem in pos_tag for token in elem]

df_pos_tags = pd.Series(list_pos_tags).value_counts().to_frame("count").reset_index()
df_pos_tags
```
